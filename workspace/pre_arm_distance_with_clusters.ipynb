```python
import pandas as pd
from ip_to_geo_1 import ip_to_geo
import json
from util import *
```

```python
df2 = pd.read_csv('period_distance_with_clusters_F20.csv', encoding='utf-8-sig')
df3 = pd.read_csv('performance_pf_F20.csv', encoding='utf-8-sig')
```

```python
df = pd.merge(df3, df2, on='User full name')
```

```python
df
```

```python
op = []
num = 0
clusters = ['High', 'Medium', 'Low', 'None']
index = {'High': 0,
        'Medium': 1,
        'Low': 2,
        'None': 3}

for i, row in df.iterrows():
    name = row['User full name']
    d = {'User full name': name}
    p = row['Performance']
    d['Passed'] = 1 if p == 'Passed' else 0
    d['Failed'] = 1 if p == 'Failed' else 0
    
    l = [0 for j in range(12)]
    one = index[row['Dedication1'].split('-')[-1]]
    two = index[row['Dedication2'].split('-')[-1]] + 4
    three = index[row['Dedication3'].split('-')[-1]] + 8
    l[one] = 1 
    l[two] = 1
    l[three] = 1
    for i in range(0,4):
        key = 'Period-I-' + clusters[i]
        d[key] = l[i]
    for i in range(0,4):
        key = 'Period-II-' + clusters[i]
        d[key] = l[i+4]
    for i in range(0,4):
        key = 'Period-III-' + clusters[i]
        d[key] = l[i+8]
    
    op.append(d)
```

```python
df_out = pd.DataFrame(op)
df_out.to_csv('./pre_arm_pf_distance_with_clusters_F20.csv', mode = 'w', index = False, header = True, encoding='utf-8-sig')
```
