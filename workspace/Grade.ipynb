```python
df = pd.read_excel('F20 Grades.xlsx')
```

```python
num = 0
for i, row in df.iterrows():
    num+=1
print(num)
```

```python
sorted_df = df.sort_values('Grade', ascending=False)
```

```python
num = 0
for i, row in sorted_df.iterrows():
    num += 1
    print(row['Grade'])
    if num == 83 or num == 166:
        print('#################################################################')
```

```python
n = 0
op = []
p = 'Good'
for i, row in sorted_df.iterrows():
    name = row['name']
    grade = row['Grade']
    n += 1
    op.append({'User full name': name, 'Grade': grade, 'Performance': p})
    if n == 83:
        p = 'Fair'
    if n == 166:
        p = 'Poor'
```

```python
df_out = pd.DataFrame(op)
df_out.to_csv('./performance_F20.csv', mode = 'w', index = False, header = True, encoding='utf-8-sig')
```

```python
n = 0
op = []
p = 'Passed'
for i, row in sorted_df.iterrows():
    name = row['name']
    grade = row['Grade']
    n += 1
    op.append({'User full name': name, 'Grade': grade, 'Performance': p})
    if n == 100:
        p = 'Failed'
```
