```python
import pandas as pd
from ip_to_geo_1 import ip_to_geo
import json
from util import *
```

```python
df1 = pd.read_csv('performance_F20.csv', encoding='utf-8-sig')
df2 = pd.read_csv('period_F20.csv', encoding='utf-8-sig')
df3 = pd.read_csv('performance_pf_F20.csv', encoding='utf-8-sig')
```

```python
df = pd.merge(df1, df2, on='User full name')
```

```python
df
```

```python
op = []
num = 0
for i, row in df.iterrows():
    name = row['User full name']
    d = {'User full name': name}
    p = row['Performance']
    d['Good'] = 1 if p == 'Good' else 0
    d['Fair'] = 1 if p == 'Fair' else 0
    d['Poor'] = 1 if p == 'Poor' else 0
    
    l = [0 for j in range(12)]
    one = int(row['Assignment 1-4'][-1])
    two = int(row['Assignment 5-8'][-1]) + 4
    three = int(row['Assignment 9-11'][-1]) + 8
    l[one] = 1 
    l[two] = 1
    l[three] = 1
    for i in range(0,4):
        key = 'Period-I-' + str(i)
        d[key] = l[i]
    for i in range(0,4):
        key = 'Period-II-' + str(i)
        d[key] = l[i+4]
    for i in range(0,4):
        key = 'Period-III-' + str(i)
        d[key] = l[i+8]
    
    op.append(d)
```

```python
df_out = pd.DataFrame(op)
df_out.to_csv('./pre_arm_F20.csv', mode = 'w', index = False, header = True, encoding='utf-8-sig')
```
